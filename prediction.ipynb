```python
import yfinance as yf
```

```python
sp500 = yf.Ticker("^GSPC")
```

```python
sp500 = sp500.history(period="max")
```

```python
sp500
```

```python
sp500.index
```

```python
sp500.plot.line(y="Close", use_index=True)
```

```python
sp500["Tomorrow"] = sp500["Close"].shift(-1)
```

```python
sp500
```

```python
sp500["Target"] = (sp500["Tomorrow"] > sp500["Close"]).astype(int)
```

```python
sp500
```

```python
sp500 = sp500.loc["1990-01-01":].copy() #use data dating from at least 1990-01-01 to obtain data similar to that of current markets
```

```python
sp500
```

```python
from sklearn.ensemble import RandomForestClassifier #radonmForestClassifier is an overfiting-resistant model, so choosing it is wise for our prediction

# estimators number of individual decision trees we want to train, the higher it is generally the better our accuracy is
# min_samples_split help to protect agains overfitting
# random_state = 1 to have that if we run the model twice we will get the same results
model = RandomForestClassifier(n_estimators=100,min_samples_split=100, random_state=1)

# We cannot use cross validation with time series data, because our results will be good for our training but horrible for the real world
n = 150
train = sp500.iloc[:-n]
test = sp500.iloc[-n:]

predictors = ["Close","Volume","Open","High","Low"]

model.fit(train[predictors],train["Target"])
```

```python
from sklearn.metrics import precision_score

# We apply our model to our test set

preds = model.predict(test[predictors])
preds

sc = 0
for i in range(len(preds)) :
    if preds[i]==test["Target"][i]:
        sc+=1
print(sc/len(preds))
```

```python
# We convert our data to a panda series data to apply our accuracy measure function
import pandas as pd
preds = pd.Series(preds, index=test.index)
```

```python
# Now we measure the accuracy of our model
precision_score(test["Target"],preds)
```

```python
combined = pd.concat([test["Target"], preds], axis= 1)
combined.plot()
```

```python
"""def predict(train, test,predictors,model):
    model.fit(train[predictors],train["Target"])
    preds=model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined=pd.concat([test["Target"],preds],axis=1)
    return combined"""

def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```

```python
def backtest(data, model, predictors, start=2500, step=250):
    """start : we take 10 years of data (nb: we have 250 days for trading a year) 2500 days and we train our model with these 10 years of data. 
    And for the step=250, means that we will take the 10 first years and predict the eleventh (11 Ã¨me) year, and after that take the eleven first 
    years and predict the twelfth year, and so on ..."""
    #all_predictions = []
    #for i in range(start, data.shape[0], step):
    #    train = data.iloc[0:i].copy()
    #    test = data.iloc[i:(i+step)].copy()
    #    predictions = predict(train,test,predictors,model)
    #    all_predictions.append(predictions)
    #    return pd.concat(all_predictions)

def backtest(data, model, predictors, start=2500, step=250):
    all_predictions = []

    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i:(i+step)].copy()
        predictions = predict(train, test, predictors, model)
        all_predictions.append(predictions)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtest(sp500,model,predictors)
```

```python
predictions["Predictions"].value_counts()
```

```python
precision_score(predictions["Target"], predictions["Predictions"])
```

```python
predictions["Target"].value_counts() / predictions.shape[0]
```

```python
horizons = [2,5,60,250,1000]
new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean()
    
    ratio_column = f"Close_Ratio_{horizon}"
    sp500[ratio_column] = sp500["Close"] / rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}"
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["Target"]
    
    new_predictors+= [ratio_column, trend_column]
```

```python
sp500 = sp500.dropna(subset=sp500.columns[sp500.columns != "Tomorrow"])
```

```python
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["Target"])
    preds = model.predict_proba(test[predictors])[:,1]
    preds[preds >=.6] = 1
    preds[preds <.6] = 0
    preds = pd.Series(preds, index=test.index, name="Predictions")
    combined = pd.concat([test["Target"], preds], axis=1)
    return combined
```
